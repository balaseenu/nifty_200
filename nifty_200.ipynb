```python
from nsetools import Nse
from nsepy import get_history
from pprint import pprint
from datetime import date
import numpy as np
import pandas as pd
import datetime
import seaborn as sns
import pywhatkit
```

```python
nifty_200 = pd.read_csv('ind_nifty200.csv')
```

```python
nifty_200.tail()
```

```python
nifty_200_lst = nifty_200['Symbol']
```

```python
value = []
start_date = datetime.date.today() - datetime.timedelta(4)
end_date = datetime.date.today()
data = pd.DataFrame()
for i in nifty_200_lst:
    print(i)
    value.append(get_history(symbol= i, start=start_date, end=end_date))
data = data.append(value,ignore_index=True,verify_integrity=False, sort=False)
```

```python
data.to_csv('nifty_200_file.txt')
```

```python
key = []
values = []
count = 0
check = 0
for comp_name in nifty_200_lst:
        for comp1 in data['Symbol']:
            if comp_name ==  comp1:
                if data['Close'][check] > data['Open'][check]:
                    check = check+1
                    count = count+1    
                else:
                    check = check+1
            else:
                if count >= 1:
                    check = check-1
                    print(data['Symbol'][check])
                    print(count)
                    key.append(data['Symbol'][check])
                    values.append(count)
                    check = check+1
                    count = 0
                else:
                    count = 0
```

```python
final = dict(zip(key,values))
```

```python
df = pd.DataFrame(list(final.items()),columns = ['Company','Bullishness out of 5']) 
```

```python
df.to_csv('df_file.csv')
```
